```python
import os
from docx import Document
from PyPDF2 import PdfFileReader
from pptx import Presentation
import csv

def convert_docx_to_txt(docx_path):
    doc = Document(docx_path)
    text_content = '\n'.join([paragraph.text for paragraph in doc.paragraphs])
    return text_content

def convert_pdf_to_txt(pdf_path):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PdfFileReader(pdf_file)
        text_content = ''
        for page_num in range(pdf_reader.numPages):
            text_content += pdf_reader.getPage(page_num).extract_text()
        return text_content

def convert_pptx_to_txt(pptx_path):
    presentation = Presentation(pptx_path)
    text_content = ''
    for slide in presentation.slides:
        for shape in slide.shapes:
            if hasattr(shape, 'text'):
                text_content += shape.text + '\n'
    return text_content

def convert_csv_to_txt(csv_path):
    with open(csv_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        text_content = '\n'.join(','.join(row) for row in csv_reader)
        return text_content

def convert_to_txt(file_path):
    _, file_extension = os.path.splitext(file_path.lower())

    if file_extension == '.docx':
        return convert_docx_to_txt(file_path)
    elif file_extension == '.pdf':
        return convert_pdf_to_txt(file_path)
    elif file_extension == '.pptx':
        return convert_pptx_to_txt(file_path)
    elif file_extension == '.csv':
        return convert_csv_to_txt(file_path)
    elif file_extension == '.txt':
        # For .txt files, just read and return the content as is
        with open(file_path, 'r', encoding='utf-8') as txt_file:
            return txt_file.read()
    else:
        raise ValueError(f"Unsupported file type: {file_extension}")

# Get a list of all files in the directory
input_directory = '/content/drive/MyDrive/MassiveDatasets'
output_directory = '/content/drive/MyDrive/MassiveDatasets2'

files_in_directory = os.listdir(input_directory)

# Iterate over each file and convert to .txt
for file_name in files_in_directory:
    file_path = os.path.join(input_directory, file_name)
    if os.path.isfile(file_path):
        text_content = convert_to_txt(file_path)

        # Check if text_content is not None before writing to a file
        if text_content is not None:
            # Save the text content to a new .txt file in the output directory
            output_file_path = os.path.join(output_directory, f'{os.path.splitext(file_name)[0]}.txt')
            with open(output_file_path, 'w', encoding='utf-8') as txt_output:
                txt_output.write(text_content)
```

```python
def similarity(doc1,doc2):
  return cosine_similarity([doc1,doc2])
```
